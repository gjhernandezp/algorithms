# I. $f(n) = \Theta(g(n))$

$$\underset{n \rightarrow \infty}{lim}\frac{f(n)}{g(n)} \le c,\ \ \ \ \ c \in \mathbb{R}^+ (0 < c < \infty)$$

implies
$$ \exists \ c_1,c_2 \in \mathbb{R}^+ ,  \ 0 < c_1 < c < c_2\ \text{ and  } \exists n_0 \in \mathbb{N} \text{ such that, } c_1 g(n) \le f(n) \le  c_2g(n)\  \forall n \in \mathbb{N} $$

## 1. example $4n^2+8 n \sin(n) = \Theta(n^2)$

$$
4n^2+8 n \sin(n)= \Theta(n^2)
$$
We check using the limit 
$$
\underset{n \rightarrow \infty}{lim} \frac{4n^2+8 n \sin(n)}{n^2} = \underset{n \rightarrow \infty}{lim} \ \frac{4n^2}{n^2}  + \frac{8 n \sin(n)}{n^2}  = \underset{n \rightarrow \infty}{lim} \ \frac{4n^2}{n^2} + \underset{n \rightarrow \infty}{lim} \ \frac{8 \sin(n)}{n} = 4 
$$

Then

$$
4n^2+8 n \sin(n)= \Theta(n^2)
$$

Since the 
$$
\underset{n \rightarrow \infty}{lim} \frac{4n^2+8 n \sin(n)}{n^2} = 4 
$$

we can use any $0 < c_1 < 4 < c_2$, for example $c_1= 3$ and $c_2 =5$, and  for $n$ suficiently large 

$$
c_1 <  \frac{4n^2+8 n \sin(n)}{n^2}  < c_2
$$

and 
$$
c_1 n^2 < 4n^2+8 n \sin(n)< c_2n^2
$$

for $c_1= 3$ and $c_2 =5$, 

$$
3n^2 < 4n^2+8 n \sin(n) <  5n^2
$$

#II. $f(n) = O(g(n))$ and $f(n) = o(g(n))$

$f(n) = O(g(n))$

$$\underset{n \rightarrow \infty}{lim}\frac{f(n)}{g(n)} \le c,\ \ \ \ \ c \in \mathbb{R}^+ (c < \infty, \text{ including } c = 0)$$

implies

$$ \exists \  c_2, c < c_2 \in \mathbb{R}^+ \text{ and  }\exists n_0 \in \mathbb{N} \text{ such that, }  f(n) \le  c_2 g(n)\  \forall n \in \mathbb{N}.$$

And

$f(n) = o(g(n))$

$$\underset{n \rightarrow \infty}{lim}\frac{f(n)}{g(n)} = 0$$

that implies that $\forall c \in \mathbb{R}^+$
$$
\underset{n \rightarrow \infty}{lim} \frac{f(n)}{g(n)}  < c
$$

We have 
$f(n) = o(g(n)) \Rightarrow f(n) = O(g(n).$



## 3. example $a n  + b   = O(n)$

$$
a n  + b  = O(n), \ \ \  a>0 
$$
We check using the limit 
$$
\underset{n \rightarrow \infty}{lim} \frac{a n  + b  }{n} = \underset{n \rightarrow \infty}{lim} \ a  + \underset{n \rightarrow \infty}{lim} \frac{b}{n}  =  a 
$$
Since the limit is $a >0 $
$$
a n  + b  = O(n)
$$

## 4. example $2n^2 = O(n^3)$

$$
2n^2 = O(n^3)
$$
We check using the limit 
$$
\underset{n \rightarrow \infty}{lim} \frac{2n^2}{n^3} = \underset{n \rightarrow \infty}{lim} \frac{2}{n}  = 0 
$$
Since the limit is $0$
$$
2n^2 = o(n^3)
$$
that implies that $\forall c \in \mathbb{R}^+$
$$
\underset{n \rightarrow \infty}{lim} \frac{2n^2}{n^3}  < c
$$
So since this holds for any $c > 0$ 
$$
2n^2 = O(n^3)
$$


## 5. example $2^n n^2 = O\left((2.1)^n)\right))$

Now apply L´Hopital to the indeterminate forms $\frac{\infty}{\infty}$
$$
\underset{x \rightarrow \infty}{lim} \frac{2^x x^2}{(2.1)^x}  = 
\underset{x \rightarrow \infty}{lim} \frac{\partial \{2^x x^2\}/\partial x}{\partial \{(2.1)^x\} /\partial x} =
\underset{x \rightarrow \infty}{lim} \frac{2^x ln(2) x^2+ 2 x 2^x }{(2.1)^x ln (2.1)} = \cdots

If $ \underset{n \rightarrow \infty}{lim}\frac{x_{n+1}}{x_n}=L<1 $

(https://math.stackexchange.com/questions/2508842/prove-that-if-lim-fracx-n1x-n-l1-then-it-converges-to-0)

#III. $f(n) = \Omega(g(n))$ and $f(n) = \omega(g(n))$

$f(n) = \Omega(g(n))$

$$c \le \underset{n \rightarrow \infty}{lim}\frac{f(n)}{g(n)} ,\ \ \ \ \ c \in \mathbb{R}^+ (c > 0, , \text{ including } c = \infty)$$

implies

$$ \exists \  c_1 \in \mathbb{R}^+ ,  0 < c_1 < c   \text{ and  }\exists n_0 \in \mathbb{N} \text{ such that, }  c_1 g(n) \le f(n) \  \forall n \in \mathbb{N}.$$

And

$f(n) = \omega(g(n))$

$$\underset{n \rightarrow \infty}{lim}\frac{f(n)}{g(n)} = \infty\ \ \ \ \ \text{or equivalently }\ \ \ \ \ \underset{n \rightarrow \infty}{lim}\frac{g(n)}{f(n)} =0$$

that implies that $\forall c \in \mathbb{R}^+$
$$
c <\underset{n \rightarrow \infty}{lim} \frac{f(n)}{g(n)}  
$$

We have 

$$f(n) = \omega(g(n)) \Rightarrow f(n) = \Omega(g(n).$$

and 

$$f(n) = \omega(g(n)) \Rightarrow g(n) = o(f(n).$$
and 

$$f(n) = o(g(n)) \Rightarrow g(n) = \omega(f(n).$$

## 7. example $\sqrt{n} = \Omega\left(lg(n)\right)$

$$
\sqrt{n} = \Omega\left(lg(n)\right)
$$
We check using the limit 
$$
\underset{n \rightarrow \infty}{lim} \frac{\sqrt{n}}{lg(n)} = \frac{\infty}{\infty}
$$
To able to use L´Hopital we change to continous fucntions (sequences do not have derivatives) 
$$
\underset{\underset{n \in \mathbb{N}}{n \rightarrow \infty}}{lim} \frac{\sqrt{n}}{lg(n)}= \underset{\underset{x \in \mathbb{R}}{x \rightarrow \infty}}{lim} \frac{\sqrt{x}}{lg(x)}
$$
Now apply L´Hopital to the indeterminate forms $\frac{\infty}{\infty}$
$$
\underset{x \rightarrow \infty}{lim} \frac{\sqrt{x}}{lg(x)}  = 
\underset{x \rightarrow \infty}{lim} \frac{\partial \{\sqrt{x}\}/\partial x}{\partial \{lg(x)\} /\partial x} =
\underset{x \rightarrow \infty}{lim} \frac{\frac{1}{2\sqrt{x}}}{\frac{1}{x}} =
\underset{x \rightarrow \infty}{lim} \frac{x}{2\sqrt{x}} = 
\underset{x \rightarrow \infty}{lim} \frac{\sqrt{x}}{2} = \infty
$$
for any constant $c>0$ and we have 
$$
c <= \underset{x \rightarrow \infty}{lim} \frac{\sqrt{x}}{lg(x)} =
\underset{n \rightarrow \infty}{lim} \frac{\sqrt{n}}{lg(n)}
$$
then
$$
c <= \underset{\underset{n \in \mathbb{N}}{n \rightarrow \infty}}{lim} \frac{\sqrt{n}}{lg(n)}, \forall c \in \mathbb{R}^+ 
$$
so 
$$
\sqrt{n} = \omega\left(lg(n)\right)
$$
that implies that $\forall c \in \mathbb{R}^+$
$$
c  < \underset{n \rightarrow \infty}{lim} \frac{\sqrt{n}}{lg(n)}  
$$
So since this holds particular $c > 0$ that implies
$$
\sqrt{n} = \Omega\left(lg(n)\right).
$$
