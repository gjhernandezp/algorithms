```python
print('Numpy version: ', np.__version__)
print('Xarray version: ', xr.__version__)
```

```python
x = np.linspace(-np.pi, np.pi, 19)
f = np.sin(x)
print(x)
print(f)
```

```python
da_f = xr.DataArray(f)
da_f
```

```python
da_f = xr.DataArray(f, dims=['x'])
da_f
```

```python
da_f = xr.DataArray(f, dims=['x'], coords={'x': x})
da_f
```

```python
da_f.plot(marker='o')
```

```python
da_f[10]
```

```python
da_f[:10]
```

```python
da_f.sel(x=0)
```

```python
da_f.sel(x=slice(0, np.pi)).plot()
```

```python
da_g = da_f**2 + 1
da_g
```

```python
da_g.plot()
```

Exercise

Multipy the DataArrays da_f and da_g together.

Select the range 

Plot the result

```python
(da_f * da_g).sel(x=slice(-1, 1)).plot(marker='o')
```
